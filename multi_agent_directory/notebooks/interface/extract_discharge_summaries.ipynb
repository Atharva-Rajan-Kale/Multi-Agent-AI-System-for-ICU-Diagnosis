```python
print("="*70)
print("📝 EXTRACTING DISCHARGE SUMMARIES")
print("="*70)

print("\nLoading NOTEEVENTS.csv (this may take 2-3 minutes)...")

# Only load the columns we need to save memory
notes_df = pd.read_csv(
    '../../data/raw/NOTEEVENTS.csv',
    usecols=['SUBJECT_ID', 'HADM_ID', 'CATEGORY', 'TEXT']
)

print(f"✅ Loaded: {notes_df.shape}")
print(f"   Total notes: {len(notes_df):,}")

# Filter for discharge summaries only
print("\nFiltering for Discharge summaries...")
discharge_summaries = notes_df[notes_df['CATEGORY'] == 'Discharge summary'].copy()

print(f"✅ Filtered: {len(discharge_summaries):,} discharge summaries")

# Drop the CATEGORY column (no longer needed)
discharge_summaries = discharge_summaries[['SUBJECT_ID', 'HADM_ID', 'TEXT']]

# Remove any rows with missing TEXT
before = len(discharge_summaries)
discharge_summaries = discharge_summaries.dropna(subset=['TEXT'])
after = len(discharge_summaries)

print(f"✅ Removed {before - after} rows with missing TEXT")
print(f"   Final: {len(discharge_summaries):,} discharge summaries with text")

# Show preview
print(f"\nPreview:")
print(discharge_summaries.head(2))
```

```python
print("="*70)
print("💾 SAVING DISCHARGE SUMMARIES")
print("="*70)

# Save to processed data folder
output_path = '../../data/processed/discharge_summaries_all.csv'

print(f"\nSaving {len(discharge_summaries):,} discharge summaries...")
print("This may take 1-2 minutes (3.7 GB file)...")

discharge_summaries.to_csv(output_path, index=False)

print(f"✅ Saved: {output_path}")
print(f"   Rows: {len(discharge_summaries):,}")
print(f"   Columns: {discharge_summaries.columns.tolist()}")

# Verify it saved correctly
saved_df = pd.read_csv(output_path, nrows=5)
print(f"\n✅ Verification - loaded first 5 rows successfully")
print(f"   Has TEXT: {'TEXT' in saved_df.columns}")
print(f"   Has SUBJECT_ID: {'SUBJECT_ID' in saved_df.columns}")

print("\n" + "="*70)
print("✅ DISCHARGE SUMMARIES SAVED!")
print("="*70)
```
